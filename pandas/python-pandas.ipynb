Making changes to the data

```python
# dataframe['Total'] = dataframe['HP'] + dataframe['Attack'] + dataframe['Defense'] + dataframe['Sp. Atk'] + dataframe['Sp. Def'] + dataframe['Speed']
# dataframe = dataframe.drop(columns=['Total'])

# criação da coluna 'Total'; selecionamos todas as linhas com ':' e as colunas de '5 a 10', pois a coluna 11 não é incluída
# então, realizamos uma soma, considerando o eixo x, ou seja, é uma soma horizontal dos valores que temos em uma linha
# obs.: começamos a contar a partir da coluna '#' (índices das linhas), cujo índice é 0, name é 1, type 1 é 2, assim por diante
dataframe['Total'] = dataframe.iloc[:, 4:10].sum(axis=1)
columns = list(dataframe.columns)
dataframe = dataframe[columns[0:4] + columns[10:12] + columns[4:10] + [columns[-1]]]
# não alteramos a tabela de fato, apenas de uma forma visual
dataframe.head(5)
```

Saving our data (exporting into desired format)

```python
# dataframe.to_csv('modified.csv')
dataframe.to_csv('modified.csv', index=False)
dataframe.to_excel('modified.xlsx', index=False)
dataframe.to_csv('modified.txt', index=False, sep='\t')
```

Working with large amounts of data

```python
# o parâmetro chunksize indica que o arquivo .csv deve ser lido em pedaços (chunks) de 100000 linhas por vez
# assim, o loop permite processar separadamente os pedaços do arquivo
# for df in pd.read_csv('modified.csv', chunksize=100000): 

# for df in pd.read_csv('modified.csv', chunksize=5): 
#   print()
#   print('CHUNK DF')
#   print(df)

new_df = pd.DataFrame(columns=dataframe.columns)
for df in pd.read_csv('modified.csv', chunksize=5):
  results = df.groupby(['Type 1']).count()
  new_df = pd.concat([new_df, results])

new_df
```
